```python
from bs4 import BeautifulSoup
import requests
```

Get the HTML of the website containing the script 

```python
website = 'https://subslikescript.com/movie/Hero_Dog_The_Journey_Home-11045566'
result = requests.get(website)
content = result.text
soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())
```

Obtain HTML code of specific elements desired using inspect

Use Beautiful Soup to obtain specific elements from the site.

```python
box = soup.find('article', class_='main-article')
```

```python
title = box.find('h1').get_text()
```

```python
transcript = box.find('div', class_='full-script')
transcript = transcript.get_text(strip=True, separator=' ')
```

```python
print(title)
```

```python
print(transcript)
```

Export data as a .text file

```python
with open(f'{title}.txt', 'w') as file:
    file.write(transcript)
```
